##How to crawl a website with Python using Advertools library


---
IF you want to have a quick round up of a new potential client's website, the following code is gonna help you save some valubale time in manual research. 
Crawl a website with the amazing Advertools library (huge shout out to Elias Dabbas for giving it the birth) and get and squezze up the package to start getting your hands dirty on the site's current Structured Data batch. 



```python
!pip install advertools
```

```python
import advertools as adv
from advertools import crawl
import pandas as pd
```

```python
site = "https://fusionunlimited.co.uk"  #@param {type:"string"}

crawl(site, 'simp.jl', follow_links=True)
crawl_df = pd.read_json('simp.jl', lines=True)
crawl_df.head()
```

```python
#@title See SEO variables included in Columns
columns = list(crawl_df) #@param {type:"string"}
```

```python
columns
```

```python
#@title Get a grip on Schema Markup Type { display-mode: "both" }
json_df = crawl_df[['url', 'jsonld_@type' ]] #@param {type:"string"}
json_df
```

```python
#@title ⭐️ Download the array above in JSON-LD
from google.colab import files

json_df.to_csv("json_df.csv")
files.download('json_df.csv')
```

```python
#@title 🔢 How many Structured Data encoded in JSON-LD ?
json_counts = json_df['jsonld_@type'].value_counts() #@param {type:"string"}

json_counts
```

```python
#@title 🤔 Which Type of Schema is About?
json_counts = json_counts.reset_index() #@param {type:"string"}
json_counts
```
